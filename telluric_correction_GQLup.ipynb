# Telluric Absorption Correction

```python
cutoff=0.6

for file in GQLup_20110620_files:
    hdulist = fits.open('./divdata/GQ_Lup/'+file)
    maxStd = 0
    maxStd = np.nanmax(hdulist[1].data['std'])
    boolArray = (hdulist[1].data['std'] < cutoff*maxStd)
    hdulist[1].data['src'][boolArray] = 'NaN'
    hdulist[1].data['div'][boolArray] = 'NaN'
    hdulist.writeto('./divdata/GQ_Lup/abs_corr_'+file,overwrite=True)
    hdulist.close()
    
for file in GQLup_20120512_files:
    hdulist = fits.open('./divdata/GQ_Lup/'+file)
    maxStd = 0
    maxStd = np.nanmax(hdulist[1].data['std'])
    boolArray = (hdulist[1].data['std'] < cutoff*maxStd)
    hdulist[1].data['src'][boolArray] = 'NaN'
    hdulist[1].data['div'][boolArray] = 'NaN'
    hdulist.writeto('./divdata/GQ_Lup/abs_corr_'+file,overwrite=True)
    hdulist.close()
    
for file in GQLup_20120702_files:
    hdulist = fits.open('./divdata/GQ_Lup/'+file)
    maxStd = 0
    maxStd = np.nanmax(hdulist[1].data['std'])
    boolArray = (hdulist[1].data['std'] < cutoff*maxStd)
    hdulist[1].data['src'][boolArray] = 'NaN'
    hdulist[1].data['div'][boolArray] = 'NaN'
    hdulist.writeto('./divdata/GQ_Lup/abs_corr_'+file,overwrite=True)
    hdulist.close()
    
for file in GQLup_20130624_files:
    hdulist = fits.open('./divdata/GQ_Lup/'+file)
    maxStd = 0
    maxStd = np.nanmax(hdulist[1].data['std'])
    boolArray = (hdulist[1].data['std'] < cutoff*maxStd)
    hdulist[1].data['src'][boolArray] = 'NaN'
    hdulist[1].data['div'][boolArray] = 'NaN'
    hdulist.writeto('./divdata/GQ_Lup/abs_corr_'+file,overwrite=True)
    hdulist.close()
    
    
    
```
